## Investigation Overview

> My goal for this project was to see the immediate effect that the 9/11 attacks had on air travel to and from New York for September of 2001.  This includes looking at total airtime, number of cancellations, departure/arrival delays, and number of total flights scheduled to and from New York for September of 2001.

> I will summarize the findings through going variable by variable.  When I refer to each Section, Section 1= 1st to 10th, Section 2= 11th to 20th, Section 3 = 21st to 30th:

This implies that it was starting to go back to pre-9/11 levels, but not quite there.

> 2. In terms of Total Airtime of all flights arriving and departing New York, it stayed at 140,000 miles or higher for Section 1.  9/11 had only morning flights, then the rest of the flights for 9/11 and 9/12 were cancelled....Airtime was extremely low until it finally surpassed 100,000 miles on the 17th.  Airtime stuck near the 100,000-115,000 level for the rest of the month, implying that less flights were scheduled as compared to Section 1. > 3.  Arrival delays peaked on the 4th and on the 10th, and overall were the heaviest in Section 1, and the middle part of Section 2 (14th= 11,067 minutes, 15th= 26,022 minutes, 16th= 16,513 minutes) which was the first weekend after 9/11. Departure delays also peaked on the 4th and on the 10th, and the middle part of Section 2 (14th-17th).  I think New York tended to have less departure delays than arrival delays for the first 10 days, and when 9/11 happened, security concerns probably caused that to flip (more departure delays than arrival) since the attacks happened in New York.

> 4.  Strangely, the two days with the most total delays (in minutes) were the 10th (68,430 minutes) and the 4th (60,936 minutes).  Total delays were highest for patches of Section 1 and that middle patch of Section 2.  Later in the month saw fewer delays partially because there were less flights.

> 5. In conclusion, it seems like 9/11 had a dramatic effect on air travel coming to and leaving from New York.  It doesn't even take into account how much longer it took passengers to go through the newly imposed security measures (how long did it take passengers from the time of arrival at the airport to actually boarding the plane?). It must have been quite dramatic to go from 1400 to 1500-ish non cancelled flights to no flights at all (the rest of 9/11 and 9/12).

Obviously, traveling by plane was never the same after 9/11...but by the end of September it seemed that levels were at least at about 75% in terms of flights and airtime.  


## Dataset Overview and Executive Summary

> This is a dataset from the Bureau of Transportation Statistics.( https://www.transtats.bts.gov/Fields.asp?gnoyr_VQ=FGJ ).  It reports flights in the United States showing  extensive arrival and departure delays from 1987 to 2022.  I decided to filter the data to be merely only New York departures and arrivals from the month of September back in 2001.  I wanted to show the consequences on air travel that the 9/11 attacks had specifically on New York during that month. 
> I used 4 different alterations of this original dataset for these visualizations presented here.  "df_delays" shows only flights that had delays of some sort (arrival, departure and both). "df_flights2" shows only flights that were not cancelled, with some outliers dropped.  "df_sections" divides the data into sums by section of month (1st-10th, 11th-20th and 21st-30th).  "df_sums" divides the data into sums by day of month (1st-30th individually).

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Loading the datasets for the visualizations
df_delays = pd.read_csv('/Users/seanward/Desktop/UDACITY/Project 4/Data Sets/df_delays.csv')
df_flights2 = pd.read_csv('/Users/seanward/Desktop/UDACITY/Project 4/Data Sets/df_flights2.csv')
df_sections = pd.read_csv('/Users/seanward/Desktop/UDACITY/Project 4/Data Sets/df_sections.csv')
df_sums = pd.read_csv('/Users/seanward/Desktop/UDACITY/Project 4/Data Sets/df_sums.csv')
```

## (Visualizations 1 and 2) Clustered Bar Charts

### The top visualization is a clustered bar count chart depicting the total number of delayed flights ordered by arrival,departure, or both...all by each section of the month.

### The second visualization is a clustered bar chart that depicts the total delays by minutes ordered by arrival, departure, or both...all by section of the month.


```python
delay = sns.countplot(data=df_delays, x='SECTION_OF_MONTH', hue_order = ['ARRIVAL', 'DEPARTURE', 'BOTH'], hue='FORM_OF_DELAY')
delay.set_title( "Total Number of Delayed Flights", fontsize= 15, fontweight='bold')
delay.set_xlabel( " ")
delay.set_ylabel( " ")
delay.legend(labels= ['ARRIVAL', 'DEPARTURE', 'BOTH'])

delay2 = df_sections.plot(x="SECTION_OF_MONTH", y=["ARR_DELAY_NEW", "DEP_DELAY_NEW", "TOTAL_DELAYS"], kind="bar",figsize=(6,3), width= 0.9)
delay2.set_title( "Total Delays by Minutes", fontsize= 15, fontweight='bold')
delay2.set_xlabel( " ")
delay2.set_ylabel( " ")
delay2.set_xticklabels(labels= ['Sept. 1-10', 'Sept. 11-20','Sept. 21-30'], rotation=0)
delay2.legend(labels= ['ARRIVAL', 'DEPARTURE', 'TOTAL'])
```

### The first visualization showing number of delayed flights shows clearly that the  number of flights with only arrival delays and with both forms of delay peaked in the first section of the month.  All three variables went down for the second section of the month, but especially arrival delays.  Number of departure flights delayed peaked in the third section of the month.

### For the second visualization showing total delays by minutes, all three variables went down from 1st to 2nd sections, and from 2nd to 3rd sections.  There was delay time with arrival flights than with departure flights in Section 1, but more departure delay time in the 2nd and 3rd sections.

### My theory is that 9/11 had a bigger effect on New York airports for departure delays because of the increased risk of further attack and the adjustment of security standards since the attacks did take place in New York City.

## (Visualization 3) Barplot
### This visualization is a barplot depicting the number of daily cancelled flights.

```python
plt.figure(figsize = [10, 9])
sns.barplot(data=df_sums, y='CANCELLED', x='DAY_OF_MONTH', color='blue')
plt.xticks(rotation=15);
plt.xlabel('September 2001', fontsize=15.0, fontweight='bold')
plt.ylabel(' ', fontsize=15.0, fontweight='bold')
plt.title('Number of Daily Cancelled Flights for September of 2001', fontsize=17.0, fontweight='bold')
plt.xlim((-1, 30));
```

## This barplot shows on a day to day basis just how dramatically 9/11 effected the number of cancelled flights.  9/10 was the highest number up to the point in the month (over 200 flights cancelled), it then shot up to 1200 on the 11th, and nearly all the flights on the 12th and 13th were cancelled.  Only the last three days showed a more normal number of cancellations.

## (Visualization 4) Scatterplot

### This visualization is a scatterplot depicting every delayed flight and how long they were delayed daily through September.

```python
fig = plt.subplots(0 ,1, figsize=(10, 8))
dom= sns.regplot(data=df_flights2, x='DAY_OF_MONTH', y='TOTAL_DELAYS', x_jitter=0.2, scatter_kws={'alpha':1/20});
dom.set_title("Total Flight Delays To and From New York for September 2001", fontsize= 15, fontweight='bold')
dom.set_xlabel( "September 2001", fontsize=16.0)
dom.set_ylabel( "Total Delays (in minutes)", fontsize=16.0)
```

## This scatterplot shows every delay for flights throughout September.  The trendline shows that the average amount of delay gradually went down during the month.  Starting around 50 and ending up possibly around 35-40 minutes per delay.  
## I can see that many of the days with the most heavy delays are the 1st, 4th, 7th-10th...then as the airports were recovering from the effects of 9/11, there were heavy delays on the 15th-17th.  
## Two of the main reasons why there were less plots towards the end of the month (in my opinion) was that the airports were starting to adjust better to the post-9/11 world...and that there were less flights scheduled overall...less flights usually means less delays.

## (Visualization 5) Clustered Bar Chart
### This is a clustered bar chart depicting the number of flights scheduled next to the number of flights cancelled for each section of the month.

```python
cancel = df_sections.plot(x="SECTION_OF_MONTH", y=["FLIGHTS", "CANCELLED"], kind="bar",figsize=(8,5), width= 0.9)
cancel.set_title( "Scheduled Flights and Cancellations (To and From New York)", fontsize= 15, fontweight='bold')
cancel.set_xlabel( "September", fontweight='bold')
cancel.set_ylabel( "Number of Flights", fontweight='bold')
cancel.set_xticklabels(labels= ['Sept. 1-10', 'Sept. 11-20','Sept. 21-30'], rotation=0, fontweight='bold')
cancel.legend(labels= ['Scheduled', 'Cancelled'])
```

## This clustered bar chart is a great way to visually see how dramatic 9/11 was in terms of cancelled flights. The first ten days (1st section) give an indication of what was "normal" for number of cancelled flights.  There was only a fraction of them, with over 14,000 flights schedule, just a handful were cancelled (629 to be exact).  There were nearly as many flights scheduled for the 2nd sections, but over half of them were cancelled (8,326!!!).  
## The 3rd section shows a far lower level than the 2nd section, but still considerably higher than the 1st section (2744).  There were also far less scheduled flights (1st: 14,618, 2nd: 14,617, 3rd: 12,686).
